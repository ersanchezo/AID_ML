```r
i = 4000 # X-dimension for bigmat
j = 4000 # Y-dimension for bigmat

bigmat = matrix(data = rnorm(n = i*j, mean = 0, 1), 
                nrow = i, ncol = j)

submat_height = 4 # subset-matrix height
submat_width  = 4 # subset-matrix width

submat_i = seq(from = 1, to = floor(i/submat_height)*submat_height, by = submat_width) # break-points along X
submat_j = seq(from = 1, to = floor(j/submat_width)*submat_width, by = submat_height)# break-points along Y
```

```r

brokemat = lapply(X = seq_along(submat_j), 
       function(j){
         j1 = submat_j[j] # Along Y, start
         j2 = submat_j[j] + (submat_height-1) # Along Y, end
         
         lapply(X = seq_along(submat_i), function(i){
           i1 = submat_i[i]
           i2 = submat_i[i] + (submat_width-1)
             
           return(bigmat[(i1:i2),(j1:j2)])
           
         })
         
       })
```

```r
#Aux function - get biggest cluster in the graph
big_clust<-function(g){
    clus<-clusters(g)
    vert_l<-which(clus$membership == which.max(clus$csize))
    ng<-induced_subgraph(g, vert_l)
    return(ng)
}
```
